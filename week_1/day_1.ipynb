```python
raghav = fetch_website_contents("https://raghav-portfolio-virid.vercel.app/")
print(raghav)
```

```python
# Define a system prompt for the model to analyze website content
system_prompt = """
You are a professional writer and content analyst. Analyze the contents of a website
and produce a clear, concise, and engaging summary that captures the key ideas and important information,
while ignoring navigation menus, headers, footers, advertisements, and other non-content elements.
Respond in markdown. Do not wrap the markdown in a code block—respond only with the markdown.
"""
```

```python
# Define a user prompt prefix for the model to analyze website content
user_prompt_prefix = """
The following is the extracted content of a website.
Analyze the content and provide a clear, professional summary that captures the website's primary purpose, key topics, products, services, or information presented.
If there are any news articles, announcements, updates, or notable events mentioned, include a brief summary of those as well.
Focus on the most relevant information and avoid repeating unnecessary details.
"""
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "What is 2 + 2?"}
]

response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
response.choices[0].message.content
```

```python
# Define a function to generate messages for the model based on the website content
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
messages_for(raghav)
```

```python
# Define a function to summarize the contents of a website using the OpenAI API
def summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
summarize("https://raghav-portfolio-virid.vercel.app/")
```

```python
# Define a function to display the summary of a website in markdown format
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://raghav-portfolio-virid.vercel.app/")
```
